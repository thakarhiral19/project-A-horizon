```python
sns.countplot(x='2urvived', data=titanic)
plt.title('Survival Count')
plt.show()
```

```python
sns.countplot(x='Sex', data=titanic)
plt.title('Gender Distribution')
plt.show()
```

```python
sns.countplot(x='Sex', hue='2urvived', data=titanic)
plt.title('Survival by Gender')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(titanic['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='2urvived', data=titanic)
plt.title('Passenger Class vs Survival')
plt.show()
```

```python
encoded = titanic.copy()

encoded['Sex'] = encoded['Sex'].map({'male':0, 'female':1})
```

```python
corr = encoded.corr(numeric_only=True)

plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
